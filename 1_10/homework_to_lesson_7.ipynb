```python
#-----ЗАДАНИЕ №1-----

class Money:    
    
    def __init__(self, rub, cop):
        self.rub = rub
        self.cop = cop
        self.check()
        
    def set_rub(self, num):
        self.rub = num

    def get_rub(self):
        return self.rub

    def set_cop(self, num):
        self.cop = num

    def get_cop(self):
        return self.cop
    
    def get_money(self):
        self.check()
        return f"{self.rub},{self.cop} рублей"
        
    def check(self): # проверка чтобы копеек не было больше 100 и меньше 0
        if self.cop > 100 or self.cop < 0:
            self.rub += self.cop // 100
            self.cop = self.cop % 100
    
    def add_money(self, summ):
        self.rub += summ.rub
        self.cop += summ.cop
        self.check()

    def sub_money(self, summ):
        self.rub -= summ.rub
        self.cop -= summ.cop
        self.check()

    def der_money(self, summ) -> float:
        return round((self.rub*100 + self.cop) / (summ.rub*100 + summ.cop), 4)

    def der_money_on_num(self, num : float):
        new_num = round((self.rub*100 + self.cop) / num)
        self.rub = int(new_num // 100)
        self.cop = int(new_num % 100)
    
    def mult_money_on_num(self, num : float):
        return self.der_money_on_num(1 / num)

    def money_is_the_sameQ(self, summ) -> bool:
        return self.rub == summ.rub and self.cop == summ.cop
```

```python
a = Money(50, 270) # изменяем по ходу выполнения программы
b = Money(80, 90) # оставляем нетронутой
print(f"первая денежная сумма: {a.get_money()}")
print(f"вторая денежная сумма: {b.get_money()}")
a.add_money(b)
print(f"увеличение первой денежной суммы на значение второй суммы: {a.get_money()}")
a.sub_money(b)
print(f"уменьшение первой денежной суммы на значение второй суммы: {a.get_money()}")
rel = a.der_money(b)
print(f"отношение (будет константным) первой суммы ко второй: {rel}")
a.der_money_on_num(rel)
print(f"деление первой суммы на их отношение: {a.get_money()}")
a.mult_money_on_num(rel)
print(f"умножение первой суммы на их отношение: {a.get_money()}")
a.der_money_on_num(rel)
print(f"еще раз поделим первую сумму на их отношение: {a.get_money()}")
print(f"две суммы равны между собой? {a.money_is_the_sameQ(b)}")
```

```python
#-----ЗАДАНИЕ №2 и 3-----

class Matrix:
    
    def __init__(self, matr):
        self.matr = matr

    def set_matrix(self, mart):
        self.matr = matr

    def get_matrix(self):
        return self.matr

    def get_num_of_strings(self):
        return len(self.matr)

    def get_num_of_columns(self):
        return len(self.matr[0])
    
    def print_matrix(self):
        for i in range(self.get_num_of_strings()):
            print(self.matr[i])

    def add_matrix(self, another_matr):
        for i in range(self.get_num_of_strings()):
            for j in range(self.get_num_of_columns()):
                self.matr[i][j] += another_matr.matr[i][j]

    def sub_matrix(self, another_matr):
        for i in range(self.get_num_of_strings()):
            for j in range(self.get_num_of_columns()):
                self.matr[i][j] -= another_matr.matr[i][j]

    def der_matrix_on_num(self, num):
        for i in range(self.get_num_of_strings()):
            for j in range(self.get_num_of_columns()):
                self.matr[i][j] = round(self.matr[i][j] / num, 1)

    def mult_matrix_on_num(self, num):
        self.der_matrix_on_num(1 / num)

class Vector(Matrix):

    def set_vector(self, vect):
        self.set_matrix(vect)

    def get_vector(self):
        return self.get_matrix()

    def get_size_of_vector(self):
        return self.get_num_of_strings()
    
    def print_vector(self):
        print(self.get_vector())

    def add_vector(self, another_vect):
        self.add_matrix(another_vect)

    def sub_vector(self, another_vect):
        self.sub_matrix(another_vect)

    def der_vector_on_num(self, num):
        self.der_matrix_on_num(num)

    def mult_vector_on_num(self, num):
        self.der_vector_on_num(1 / num)
        
    def mult_matrix_on_vector(self, matrix : Matrix):
        new_list : list = []
        for i in range(matrix.get_num_of_strings()):
            summ = 0
            for j in range(matrix.get_num_of_columns()):
                summ += (matrix.matr[i][j] * self.get_vector()[j])    
            new_list.append(summ) 
        new_vector = Vector(new_list)
        return new_vector
```
