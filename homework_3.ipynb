```python
import math as m

def find_y(x): 
    y : float = (m.cos(m.exp(x)) + m.log((1 + x)**2) + m.sqrt(m.exp(m.cos(x)) + (m.sin(m.pi*x))**2) + m.sqrt(1/x) + m.cos(x**2))**m.sin(x)
    print(f"y для x = {x}: ", round(y,5))
x : int = 1.79
print("задание 1")
find_y(x)
```

```python
def search_substr(subst, st):
    if len(st) > len(subst): # проверка чтобы подстрока не была длиннее строки
        st, subst = st.casefold(), subst.casefold() # приводим все символы в обеих строках к нижнему регистру
        if st.find(subst) != -1: # проверяем наличие subst в st. если есть - возвращает индект символа, с которого начинается совпадение, если нет - возвращает -1
            print("Есть контакт")
        else:
            print("Мимо")
    else:
        print("Error")
stroka : str = "Hello world"
substroka : str = "hello"
print("задание 2")
print(f"Для строки {stroka} и подстроки {substroka}")
search_substr(substroka, stroka)
```

```python
def max_count_three(st) -> None: # функция по условию задания
    new_set : list = list(set(st)) # получаем список из уникальных символов при помощи преобразования типов в set и обратно
    list_of_count : list = [] # в этом списке будет лежать количества повторений уникальных символов в исходной строке
    
    try:
        new_set.remove(' ') # убираем пробелы
    except ValueError:
        print("нет пробелов")
    
    for i in range(len(new_set)):
        list_of_count.append(st.count(new_set[i]))
    for i in range(1, 4):
        max_count(new_set, list_of_count, i) # в данной функции находим наиболее повторяющийся символ, выводим его и удаляем его из списков new_set и list_of_count, i нужно для print

def max_count(n_set : list, list_of_c: list, i : int) -> None:
    max_count : int = max(list_of_c)
    position_of_max_count : int = list_of_c.index(max_count)
    print(f"{i}: максимально встречающийся символ (встречается {max_count} раза): {n_set[position_of_max_count]}")
    n_set.pop(position_of_max_count)
    list_of_c.pop(position_of_max_count)
    
stroka : str = "Hello world! My name is Nikita"
print("задание 3")
print(f"Для строки {stroka}")
max_count_three(stroka)
```

```python
# функции по шаблону из предыдущего задания
def count_it(st) -> dict: 
    new_set : list = list(set(st)) 
    list_of_count : list = []
    our_dict : dict = {}
    for i in range(len(new_set)):
        list_of_count.append(st.count(new_set[i]))
    for i in range(1, 4):
        res = max_count(new_set, list_of_count)
        our_dict.setdefault(res[0], res[1]) # если ключа не существует (а он не существует), то добавляет пару ключ-значение
    return our_dict

def max_count(n_set : list, list_of_c: list) -> int:
    max_count : int = max(list_of_c)
    position_of_max_count : int = list_of_c.index(max_count)
    value = n_set[position_of_max_count]
    n_set.pop(position_of_max_count)
    list_of_c.pop(position_of_max_count)
    return int(value), max_count
    
stroka : str = "12233344445678999999"
print("задание 4")
print(f"Для строки {stroka}")
print(f"Наш словарь: {count_it(stroka)}")
```
