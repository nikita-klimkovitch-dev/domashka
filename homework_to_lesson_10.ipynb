```python
class Delivery:
    
    def __init__(self, transport, distance, weight):
        self.transport = transport
        self.distance = distance
        self.weight = weight
                        
    def calculate_price_and_speed_of_the_transport(self)-> list: 
        if self.transport == "car": 
            return [10, 70]
        elif self.transport == "track": 
            return [20, 100]
        elif self.transport == "train": 
            return [15, 50]
        elif self.transport == "plane": 
            return [50, 700]
        elif self.transport == "ship": 
            return [17, 30] 
     
    def calculate_time(self):
        return round(self.distance / self.calculate_price_and_speed_of_the_transport()[1], 2)
        
    def calculate_price(self):
        return self.distance * self.calculate_price_and_speed_of_the_transport()[0] * self.weight   

    def print_info(self):
        print("Your delivery time:",  self.calculate_time(), "h, Your delivery price:", self.calculate_price(), "rub")
        
```

```python
del1 = Delivery("train", 1000, 2)
del1.print_info()
```

```python
class Express_Delivery(Delivery):
    
    def __init__(self, distance, weight):
        super().__init__("track", distance, weight)

    def calculate_price(self):
        return self.distance * self.calculate_price_and_speed_of_the_transport()[0] * self.weight * 1.2
        
    def calculate_time(self):
        return self.distance / self.calculate_price_and_speed_of_the_transport()[1] * 0.7   
```

```python
del2 = Express_Delivery(1000, 2)
del2.print_info()
```

```python
class Premium_Delivery(Express_Delivery):
    
     def __init__(self, distance, weight):
        super().__init__(distance, weight)
                 
     def calculate_price(self):
        return self.distance * self.calculate_price_and_speed_of_the_transport()[0] * self.weight * 1.3

     def print_info(self):
        print("Your delivery time:",  self.calculate_time(), "h, Your delivery price:", self.calculate_price(), "rub")
        print("Your bonus:", self.calculate_price()/100, f"points, Your tracking number: {self.distance}RTFTGER")      
```

```python
del3 = Premium_Delivery(1000, 2)
del3.print_info()
```

```python
class International_Delivery(Delivery):
    
     def __init__(self, distance, weight, product_price):
        super().__init__("ship", distance, weight)
        self.product_price = product_price
                 
     def calculate_price(self):
        return self.distance * self.calculate_price_and_speed_of_the_transport()[0] * self.weight + self.product_price * 0.2
    
     def print_info(self):
        print("Your delivery time:",  self.calculate_time(), "h, Your delivery price:", self.calculate_price(), "rub")
        print("Your custom fee:", self.product_price * 0.2)         
```

```python
del4 = International_Delivery(1000, 2, 10000)
del4.print_info()
```

```python
class Free_Delivery(Delivery):
     def __init__(self, distance, weight, product_price):
        super().__init__("car", distance, weight)
        self.product_price = product_price               
    
     def print_info(self):
        if self.distance < 100 and self.product_price >= 10000: 
            print("Your delivery time:", self.calculate_time(), "h, Your delivery price:", 0, "rub")
        else:
            print("Your delivery isn't free")
            super().print_info()
```

```python
del5 = Free_Delivery(200, 2, 10000)
del5.print_info()
```
