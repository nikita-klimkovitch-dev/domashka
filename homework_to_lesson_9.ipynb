```python
import random

class Array:

    def __init__(self, size, method):
        if method == 'custom':
            self.arr = []
            for i in range(size):
                a = input()
                self.arr.append(int(a))
        if method == 'random':
            self.arr = [random.randint(1, 100) for _ in range(size)]

    def set_arr(self, arr):
        self.arr = arr

    def get_arr(self):
        return self.arr

    def print_arr(self):
        print(self.arr)

    def get_size(self):
        return len(self.arr)

    def total_of_elements(self):
        summ = 0
        for i in range(self.get_size()):
            summ += self.arr[i]
        return summ

    def add_arr(self, new_arr):
        for i in range(self.get_size()):
            self.arr[i] += new_arr.get_arr()[i]
```

```python
first_arr = Array(4, 'random')
print('первый массив: ')
first_arr.print_arr()
print('сумма элементов в массиве: ', first_arr.total_of_elements())
print('второй массив: ')
second_arr = Array(4, 'custom')
second_arr.print_arr()
print('увеличение значений элементов первого массива на значения элементов второго массива: ')
first_arr.add_arr(second_arr)
first_arr.print_arr()
```
