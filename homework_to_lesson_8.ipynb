```python
print("Task 1")

import math

class Shape:
    
    def area(self):
        pass

    def perimeter(self):
        pass

    def display_info(self):
        self.get_info()

    def get_name(self):
        print(f"тип фигуры {self.__class__.__name__}") # получение "чистого" имени класса

class Circle(Shape):

    def __init__(self, radius):
        self.set_radius(radius)
  
    def set_radius(self, radius):
        if radius > 0:
            self.radius = radius
        else:
            print("Радиус должен быть строго больше нуля")
            self.radius = 0
    
    def get_radius(self):
        return self.radius

    def area(self):
        return round(3.14 * self.radius**2, 2)
        
    def perimeter(self):
        return round(2 * 3.14 * self.radius, 2)
        
    def get_info(self):
        print(f"радиус {self.radius}, площадь {self.area()}, периметр {self.perimeter()}")

class Rectangle(Shape):

    def __init__(self, length, height):
        self.set_length(length)
        self.set_height(height)
  
    def set_length(self, length):
        if length > 0:
            self.length = length
        else:
            print("Длина должна быть строго больше нуля")
            self.length = 0

    def set_height(self, height):
        if height > 0:
            self.height = height
        else:
            print("Высота должга быть строго больше нуля")
            self.height = 0    
            
    def get_length(self):
        return self.length

    def get_height(self):
        return self.height

    def area(self):
        return round(self.length * self.height, 2)
        
    def perimeter(self):
        return round(2 * (self.length + self.height), 2)
        
    def get_info(self):
        print(f"длина {self.length}, высота {self.height}, площадь {self.area()}, периметр {self.perimeter()}")

class Triangle(Shape):

    def __init__(self, side_a, side_b, side_c):
        self.side_a, self.side_b, self.side_c = 0, 0, 0
        self.set_side_a(side_a)
        self.set_side_b(side_b)
        self.set_side_c(side_c)

    def check_trig(self, side) -> bool:
        if self.side_a != 0 and self.side_b != 0 and self.side_c != 0:
            if not (self.side_a + self.side_b > self.side_c and self.side_b + self.side_c > self.side_a and self.side_a + self.side_c > self.side_b):
                print("не выполняется условие существования треугольника")
                return False
            else:
                return True
        return True
    
    def set_side_a(self, side_a):
        if side_a > 0:
            self.side_a = side_a
        else:
            print("Сторона 'a' должна быть строго больше нуля")
            self.side_a = 0
        if not self.check_trig(self.side_a):
            self.side_a = 0
    
    def set_side_b(self, side_b):
        if side_b > 0:
            self.side_b = side_b
        else:
            print("Сторона 'b' должна быть строго больше нуля")
            self.side_b = 0
        if not self.check_trig(self.side_b):
            self.side_b = 0

    def set_side_c(self, side_c):
        if side_c > 0:
            self.side_c = side_c
        else:
            print("Сторона 'c' должна быть строго больше нуля")
            self.side_c = 0
        if not self.check_trig(self.side_c):
            self.side_c = 0
            
    def get_side_a(self):
        return self.side_a

    def get_side_b(self):
        return self.side_b

    def get_side_c(self):
        return self.side_c

    def area(self):
        p = self.perimeter() / 2
        if self.side_a != 0 and self.side_b != 0 and self.side_c != 0:
            return round(math.sqrt(p * (p - self.side_a) * (p - self.side_b) * (p - self.side_c)), 2)
        
    def perimeter(self):
        return round(self.side_a + self.side_b + self.side_c, 2)
        
    def get_info(self):
        print(f"сторона 'а' {self.side_a}, сторона 'b' {self.side_b}, сторона 'c' {self.side_c}, площадь {self.area()}, периметр {self.perimeter()}")

class Square(Rectangle):

    def __init__(self, side):
        super().__init__(side, side)
  
    def set_side(self, side):
        if side > 0:
            self.length = side
            self.height = side
        else:
            print("сторона квадрата должна быть строго больше нуля")
            self.length = 0
            self.height = 0
         
    def get_side(self):
        return self.length
        
    def get_info(self):
        print(f"сторона {self.get_side()}, площадь {self.area()}, периметр {self.perimeter()}")
```

```python
print("---окружность---")
circ = Circle(5)
circ.get_name()
circ.display_info()
print("---прямоугольник---")
rect = Rectangle(5, 6)
rect.get_name()
rect.display_info()
print("---треугольник---")
trig = Triangle(3, 4, 6)
trig.get_name()
trig.display_info()
trig.set_side_a(100)
trig.display_info()
print("---квадрат---")
sq = Square(5)
sq.get_name()
sq.display_info()
```

```python
print("Task 2")

class Employee:

    def __init__(self, ID, name, position, seniority, num_of_use_rest_days):
        self.ID = ID
        self.name = name
        self.position = position
        self.seniority = seniority
        self.num_of_use_rest_days = num_of_use_rest_days

    def get_ID(self):
        return self.ID

    def get_name(self):
        return self.name

    def set_position(self, position):
        self.position = position

    def get_position(self):
        return self.position

    def set_seniority(self, seniority):
        self.seniority = seniority

    def get_seniority(self):
        return self.seniority

    def set_num_of_use_rest_days(self, num):
        self.num_of_use_rest_days = num

    def get_num_of_use_rest_days(self):
        return self.num_of_use_rest_days

    def get_holiday_money(self):
        max_num_of_holiday_days = 24
        return round((max_num_of_holiday_days - self.num_of_use_rest_days) * self.get_pure_salary() / 30, 2)

    def get_info(self):
        print(f'сотрудник {self.name} с ID {self.ID}, на позиции {self.position} со стажем {self.seniority} лет (в этом месяце заработная плата {self.get_salary()} руб) за год сходил в отпуск на {self.num_of_use_rest_days} дней (оставшиеся отпускные {self.get_holiday_money()} руб)')

class PartTimeEmployee(Employee):
    
    def __init__(self, ID, name, position, seniority, num_of_use_rest_days, num_of_hours, base_salary_per_hour, benefits):
        super().__init__(ID, name, position, seniority, num_of_use_rest_days)
        self.num_of_hours = num_of_hours
        self.base_salary_per_hour = base_salary_per_hour
        self.benefits = benefits

    def set_num_of_hours(self, num_of_hours):
        self.num_of_hours = num_of_hours

    def get_num_of_hours(self):
        return self.num_of_hours

    def get_base_salary_per_hour(self):
        return self.base_salary_per_hour

    def get_benefits(self):
        return self.benefits

    def get_pure_salary(self):
        return round((8 * 24 * self.base_salary_per_hour) * (1 + self.seniority / 5) * 0.87, 2) # 13% - подоходный налог, 8 * 24 - 8 часов в течении 24 дней

    def get_salary(self):
        return round((self.num_of_hours * self.base_salary_per_hour) * (1 + self.seniority / 5) * 0.87, 2)

    def get_info(self):
        super().get_info()
        print(f'за месяц отработал {self.num_of_hours} часов c оплатой за час работы {self.base_salary_per_hour} руб, плюшка - {self.benefits}')

class CommissionEmployee(Employee):
    
    def __init__(self, ID, name, position, seniority, num_of_use_rest_days, num_of_sick_days, base_salary, volume_of_sale, percent_of_plan):
        super().__init__(ID, name, position, seniority, num_of_use_rest_days)
        self.num_of_sick_days = num_of_sick_days
        self.base_salary = base_salary
        self.volume_of_sale = volume_of_sale
        self.percent_of_plan = percent_of_plan

    def set_num_of_sick_days(self, num):
        self.num_of_sick_days = num

    def get_num_of_sick_days(self):
        return self.num_of_sick_days

    def get_base_salary(self):
        return self.base_salary

    def set_volume_of_sale(self, volume_of_sale):
        self.volume_of_sale = volume_of_sale

    def get_volume_of_sale(self):
        return self.volume_of_sale

    def set_percent_of_plan(self, percent_of_plan):
        self.percent_of_plan = percent_of_plan

    def get_percent_of_plan(self):
        return self.percent_of_plan
        
    def get_pure_salary(self):
        return round(self.base_salary * (1 + self.seniority / 5) * 0.87, 2)
    
    def get_salary(self):
        if self.percent_of_plan <= 100:
            return round(self.get_pure_salary() * ((30 - self.num_of_sick_days) / 30) + 0.05 * self.volume_of_sale, 2) # надбавка 5% от объема продаж
        else:
            return round((self.get_pure_salary() * ((30 - self.num_of_sick_days) / 30) + 0.05 * self.volume_of_sale) * 1.2, 2) # надбавка 20% если план перевыполнен
        
    def get_info(self):
        super().get_info()
        print(f'проболел {self.num_of_sick_days} дней, базавая зарплата в месяц {self.base_salary} руб, объем продаж {self.volume_of_sale} руб, план выполнен на {self.percent_of_plan} %')

class FullTimeEmployee(Employee):
    
    def __init__(self, ID, name, position, seniority, num_of_use_rest_days, num_of_sick_days, base_salary, benefits, month):
        super().__init__(ID, name, position, seniority, num_of_use_rest_days)
        self.num_of_sick_days = num_of_sick_days
        self.base_salary = base_salary
        self.benefits = benefits
        self.month = month

    def set_num_of_sick_days(self, num):
        self.num_of_sick_days = num

    def get_num_of_sick_days(self):
        return self.num_of_sick_days

    def get_base_salary(self):
        return self.base_salary

    def get_benefits(self):
        return self.benefits
        
    def set_month(self, month):
        self.month = month

    def get_month(self):
        return self.month
        
    def get_pure_salary(self):
        return round(self.base_salary * (1 + self.seniority / 5) * 0.87, 2)
    
    def get_salary(self):
        if self.month == 12:
            return round(self.get_pure_salary() * ((30 - self.num_of_sick_days) / 30) * 2, 2) # в декабре 13-ая зарплата
        else:
            return round(self.get_pure_salary() * ((30 - self.num_of_sick_days) / 30), 2)
        
    def get_info(self):
        super().get_info()
        print(f'проболел {self.num_of_sick_days} дней, базавая зарплата в месяц {self.base_salary} руб, текущий месяц № {self.month}, плюшка - {self.benefits}')

class Manager(FullTimeEmployee):

    def __init__(self, ID, name, position, seniority, num_of_use_rest_days, num_of_sick_days, base_salary, benefits, month, control_bonus, money_for_studying):
        super().__init__(ID, name, position, seniority, num_of_use_rest_days, num_of_sick_days, base_salary, benefits, month)
        self.control_bonus = control_bonus
        self.money_for_studying = money_for_studying

    def get_control_bonus(self): # коэффициент
        return self.control_bonus
    
    def set_money_for_studying(self, num):
        self.money_for_studying = num
    
    def get_money_for_studying(self):
        return self.money_for_studying
        
    def get_pure_salary(self):
        return round(self.base_salary * (1 + self.seniority / 5) * self.control_bonus * 0.87, 2)
        
    def get_info(self):
        super().get_info()
        print(f'бонус за управление командой {self.control_bonus}, бюджет на ее обучения {self.money_for_studying} руб')

class Department:

    def __init__(self, name, array_of_fulltimeemployees, array_of_parttimeemployees, array_of_commissionemployees, array_of_managers):
        self.name = name
        self.array_of_fulltimeemployees = array_of_fulltimeemployees
        self.array_of_parttimeemployees = array_of_parttimeemployees
        self.array_of_commissionemployees = array_of_commissionemployees
        self.array_of_managers = array_of_managers

    def set_name(self, name):
        self.name = name

    def get_name(self):
        return self.name
    
    def add_fulltimeemployee(self, person : FullTimeEmployee):
        self.array_of_fulltimeemployees.append(person)

    def add_parttimeemployee(self, person : PartTimeEmployee):
        self.array_of_parttimeemployees.append(person)

    def add_commissionemployee(self, person : CommissionEmployee):
        self.array_of_commissionemployees.append(person)

    def add_manager(self, person : Manager):
        self.array_of_managers.append(person)

    def get_fulltimeemployees(self):
        for i in range(len(self.array_of_fulltimeemployees)):
            print(self.array_of_fulltimeemployees[i].get_name())

    def get_parttimeemployees(self):
        for i in range(len(self.array_of_parttimeemployees)):
            print(self.array_of_parttimeemployees[i].get_name())

    def get_commissionemployees(self):
        for i in range(len(self.array_of_commissionemployees)):
            print(self.array_of_commissionemployees[i].get_name())

    def get_managers(self):
        for i in range(len(self.array_of_managers)):
            print(self.array_of_managers[i].get_name())

    def generate_report(self):
        print(f'в отделе {self.name} на полную ставку работает {len(self.array_of_fulltimeemployees)}, на неполную {len(self.array_of_parttimeemployees)}, в комиссии {len(self.array_of_commissionemployees)}, менеджеров {len(self.array_of_managers)}')
```
